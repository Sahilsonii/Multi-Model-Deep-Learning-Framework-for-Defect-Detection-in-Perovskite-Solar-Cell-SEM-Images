# Custom Dataset Training

```python
from ultralytics import YOLO
```

```python
# Load the YOLO11 model
model = YOLO("yolov8s-cls.pt")  # 'n' denotes the nano version

# Train the model
model.train(data="dataset", epochs=50, imgsz=224, device=0, batch=16)
```
